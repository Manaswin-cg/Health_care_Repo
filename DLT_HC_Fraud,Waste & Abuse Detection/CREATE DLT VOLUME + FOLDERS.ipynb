```python
spark.sql("""
CREATE CATALOG IF NOT EXISTS DLT_HC_Fraud_detection
""")

print("Catalog ready")
```

```python
spark.sql("""
CREATE SCHEMA IF NOT EXISTS DLT_HC_Fraud_detection.bronze_silver_gold
""")

print("Schema ready")
```

```python
spark.sql("""
CREATE VOLUME IF NOT EXISTS DLT_HC_Fraud_detection.bronze_silver_gold.healthcare_fraud_dlt
""")

print("Volume ready")
```

```python
catalog = "DLT_HC_Fraud_detection"
schema  = "bronze_silver_gold"
volume  = "healthcare_fraud_dlt"

base_path = f"/Volumes/{catalog}/{schema}/{volume}"

print("DLT base path:", base_path)
```

```python
folders = [
    "HC_2026_claims",
    "HC_2026_providers",
    "HC_2026_payments",
    "HC_2026_patients",
    "HC_2026_claim_lines",
    "HC_2026_rules",
    "HC_2026_audit_logs"
]

for f in folders:
    dbutils.fs.mkdirs(f"{base_path}/{f}")
    print("Created:", f"{base_path}/{f}")
```

```python
old_base = "/Volumes/health_care_2026/healthcare_fraud_demo/healthcare_fraud"

for f in folders:
    dbutils.fs.cp(
        f"{old_base}/{f}",
        f"{base_path}/{f}",
        True
    )
    print("Copied:", f)
```
