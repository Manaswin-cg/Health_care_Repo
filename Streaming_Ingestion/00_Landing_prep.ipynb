# 00_Landing_prep – Raw Source Landing Zone

The **Landing Layer** represents the initial entry point of data into the healthcare lakehouse.

This layer stores raw files exactly as received from upstream systems before any streaming ingestion or transformation occurs.

It acts as a **controlled staging area** between external data sources and the lakehouse ingestion pipelines.

---

# Purpose of the Landing Layer

- Receive raw healthcare data files from upstream systems
- Preserve original file structure and contents
- Provide a stable handoff point for Auto Loader streaming jobs
- Allow validation and replay of source files
- Enable traceability of file-based data ingestion

---

# Landing Tables by Source Domain

These tables represent raw datasets received from operational systems:

### `providers`
Contains raw hospital and doctor master data.

### `patients`
Contains patient demographic information in source format.

### `payers`
Contains insurance company master data.

### `policies_nested`
Contains policy records in original nested JSON structure.

### `claims`
Contains claim header data exactly as received.

### `claim_lines`
Contains procedure-level claim data in nested format.

### `payments`
Contains insurer payment records in source structure.

---

# Role of Landing in the Pipeline

Landing ensures:

- Files are available before ingestion begins
- Data can be replayed if ingestion fails
- File integrity can be validated
- Raw source truth is preserved for compliance

---

# Why Landing Layer Matters in Healthcare Systems

Healthcare pipelines often require:

- Data lineage tracking
- File-level audit history
- Source reprocessing ability
- Regulatory traceability

The Landing layer provides the **original evidence of incoming data**, which is critical for audits and compliance reviews.

---

# Business Value of Landing Layer

-Ensures no data loss before ingestion  
-Allows replay of source files for corrections  
-Supports audit and compliance requirements  
-Provides controlled input boundary for streaming jobs  
-Separates external data issues from internal processing  

---

The Landing layer guarantees that **all source healthcare data enters the platform safely before being transformed into analytics-ready datasets.**
